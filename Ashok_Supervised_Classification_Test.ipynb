```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
from scipy.stats import zscore
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('Absenteeism_at_work.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df['Absenteeism'].value_counts()
```

```python
df=df.drop('ID',axis=1)
```

```python
df.describe().T
```

```python
for feature in df.columns: 
    if df[feature].dtype == 'object':
        df[feature] = pd.Categorical(df[feature]).codes 
```

```python
df.info()
```

```python
train_char_label = ['0','1','2'] 
```

```python
df['Absenteeism'].value_counts() 
# 0 - High 
# 1 - low
# 2 - medium
```

```python
df.corr()
```

```python
from sklearn.tree import DecisionTreeClassifier 
```

```python
X =  df.drop("Absenteeism", axis=1)
y =  df.pop("Absenteeism")
```

```python
X.head()
```

```python
train_set, test_set, train_labels, test_labels = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
dt_model = DecisionTreeClassifier(criterion = 'entropy' )
```

```python
dt_model.fit(train_set, train_labels)
```

```python
dt_model.score(test_set , test_labels) 
```

```python
predicted_test_labels = dt_model.predict(test_set)
```

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(test_labels,predicted_test_labels)

conf_matrix
```

```python
print(metrics.classification_report(test_labels, predicted_test_labels))
```

```python
from sklearn.metrics import accuracy_score,cohen_kappa_score
cohen_kappa_score(test_labels, predicted_test_labels)
```

```python
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
```

```python
lr=LogisticRegression()
rf=RandomForestClassifier()
knn=KNeighborsClassifier()
dt=DecisionTreeClassifier()
nb=GaussianNB()
models = [lr, nb, knn, dt, rf]
for mod in models : 
    print(mod)
    print("\n")
    mod.fit(X,y)
    print(mod.score(X,y))
    ypred = mod.predict(X)
    mod.score(X , y) 
    conf_matrix = metrics.confusion_matrix(y, ypred)
    print(conf_matrix)
    print(metrics.classification_report(y, ypred))
    print("\n \n")
```
